```python
import pandas as pd
import numpy as np
# reading the CSV file and storing it to dataframe 'raw_data'
raw_data = pd.read_csv('../data-raw/raw_data.csv')
# finding columns where data in na
raw_data.columns[raw_data.isna().any()].tolist()
```

```python
# checking number of missing values for Mileage
print(raw_data['Mileage'].isna().value_counts())

# removing 2 rows are they are less compared to overall datasize
raw_data = raw_data[raw_data['Mileage'].notna()]

# checking number of missing values for Mileage
print(raw_data['Engine'].isna().value_counts())

#removing 36 rows are they are less compared to overall datasize
raw_data = raw_data[raw_data['Engine'].notna()]

# checking number of missing values for Power
print(raw_data['Power'].isna().value_counts())

# checking number of missing values for Seats
print(raw_data['Seats'].isna().value_counts())

#drop those 2 rows 
raw_data = raw_data[raw_data['Seats'].notna()]

#checking number of missing values for New_Price
print(raw_data['New_Price'].isnull().value_counts())

raw_data['New_Price'] = raw_data['New_Price'].str.replace(" Lakh", "").str.replace(" Cr","").astype(float)

new_price_median = raw_data[raw_data['New_Price'].notna()]['New_Price'].median()

print(new_price_median)

#replacing NA values with median 
raw_data['New_Price'] = raw_data['New_Price'].fillna(value = new_price_median)

raw_data.head()
#print(raw_data['New_Price'].isnull().value_counts())
```

```python
# removing extra atrributes and keeping only numerical values
raw_data['Year_Transform'] = raw_data['Year'].astype(int)
raw_data['Mileage'] = raw_data['Mileage'].str.replace(" kmpl","").str.replace(" km/kg","").astype(float)
raw_data['Engine'] = raw_data['Engine'].str.replace(" CC","").astype(int)
raw_data['Power'] = raw_data['Power'].str.replace(" bhp","").astype(float)

raw_data.head()
```

```python
raw_data['Fuel_Type'].value_counts()
raw_data['Transmission'].value_counts()

# one hot encoding for Fuel_Type and Transmission
raw_data = pd.get_dummies(
    data=raw_data,
    columns=["Fuel_Type", "Transmission"],
    prefix=["Fuel_Type", "Transmission"])

raw_data.head()
```

```python
# calculating current age of car

raw_data['Current_Age_In_Years'] = 2023 - raw_data['Year_Transform']

raw_data.head()
```
